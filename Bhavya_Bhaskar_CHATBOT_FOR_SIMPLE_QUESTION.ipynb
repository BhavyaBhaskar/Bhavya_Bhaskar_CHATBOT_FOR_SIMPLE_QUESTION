```python
def add_new_qa():
    """Function to add new question-answer pair."""
    print("\nðŸ“ Add a new question and answer")
    new_q = input("Enter your new question: ").strip().lower()
    new_a = input("Enter the answer for this question: ").strip()
    if new_q in responses:
        print("âš ï¸ This question already exists in the database.")
    else:
        responses[new_q] = [new_a]
        print("âœ… New question-answer pair added successfully!\n")

def chatbot():
    print("ðŸ¤– Hello! I'm a simple chatbot. Ask me a question (type 'exit' to quit, 'add' to add new Q&A).")
    while True:
        user_input = input("You: ").strip().lower()

        if user_input == "exit":
            print("ðŸ¤– Goodbye! Have a great day!")
            break
        elif user_input == "add":
            add_new_qa()
        else:
            response = responses.get(user_input, ["Sorry, I don't understand that question."])
            print(f"Bot: {response[0]}")

# Run the chatbot
chatbot()
```
